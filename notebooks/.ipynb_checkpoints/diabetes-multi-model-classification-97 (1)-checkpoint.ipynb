### Checking for Outliers using **Interquartile Range (IQR) method**

**The IQR (Interquartile Range)** method helps detect outliers by using the spread of the middle 50% of your data.


---


2. **Q1 (25th percentile):**
The value below which 25% of the data lies.

3. **Q3 (75th percentile):**
The value below which 75% of the data lies.

4. **IQR:**
The difference between Q3 and Q1.
$$
   IQR = Q3 - Q1
   $$
---

### Outlier Detection:
- **Lower bound**:  
   $$
   \text{Lower bound} = Q1 - 1.5 \times IQR
   $$

- **Upper bound**:  
   $$
   \text{Upper bound} = Q3 + 1.5 \times IQR
   $$

Any value **below the lower bound** or **above the upper bound** is considered an **outlier**.


---


### Summary:
The IQR method identifies where the majority of data points lie and flags any values that fall too far outside this range as outliers.


#### IQR Function

````python
def detect_outliers_iqr(df, numerical_columns):
    outliers = pd.DataFrame()
    for col in numerical_columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1

        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        col_outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]

        if not col_outliers.empty:
            outliers = pd.concat([outliers, col_outliers])

    # Remove duplicate outlier rows
    outliers = outliers.drop_duplicates()
    return outliers
````

#### Applying the outlier removal function to each class on its own

````python
# Separate the classes
df_0 = train_df[train_df['diabetes'] == 0].copy()
df_1 = train_df[train_df['diabetes'] == 1].copy()

# Define numerical columns
numerical_cols = list(X_train.select_dtypes(include=['float64']).columns)
if 'blood_glucose_level' not in numerical_cols:
    numerical_cols.append('blood_glucose_level')

# Detect and remove outliers for each class
outliers_0 = detect_outliers_iqr(df_0, numerical_cols)
outliers_1 = detect_outliers_iqr(df_1, numerical_cols)

df_0_clean = df_0.drop(outliers_0.index)
df_1_clean = df_1.drop(outliers_1.index)

# Combine cleaned training data
train_clean = pd.concat([df_0_clean, df_1_clean], ignore_index=True)

# Split back into features and labels
X_train_clean = train_clean.drop('diabetes', axis=1)
y_train_clean = train_clean['diabetes']

# Optional: Check class balance
print(y_train_clean.value_counts())
````

### Scaling our numerical data

### 🔧 Feature Scaling in Machine Learning

In machine learning, **feature scaling** is a crucial preprocessing step that transforms numerical features to a common scale without distorting differences in the ranges of values. This ensures that no single feature dominates others due to its scale, leading to more balanced and accurate models.

---

### 📈 Why Is Feature Scaling Important?

- **Equal Contribution**: Features with larger scales can unduly influence the model's predictions. Scaling ensures each feature contributes equally.
- **Improved Convergence**: Algorithms like gradient descent converge faster when features are on similar scales.
- **Enhanced Performance**: Distance-based algorithms (e.g., KNN, SVM) perform better with scaled data.

---

### ⚙️ Common Scaling Techniques

#### 1. Standardization (Z-score Normalization)
- **Description**: Centers features by removing the mean and scales them to unit variance.
- **Formula**:  
  $$[
  X_{\text{scaled}} = \frac{X - \mu}{\sigma}
  ]$$
- **Use When**:
  - Data follows a Gaussian distribution.
  - Algorithms assume standardized data (e.g., Logistic Regression, SVM).

#### 2. Min-Max Scaling (Normalization)
- **Description**: Scales features to a fixed range, typically [0, 1].
- **Formula**:  
  $$ [
  X_{\text{scaled}} = \frac{X - X_{\min}}{X_{\max} - X_{\min}}
  ] $$
- **Use When**:
  - You need a bounded range.
  - Algorithms like KNN or Neural Networks are used.

#### 3. Robust Scaling
- **Description**: Uses median and interquartile range, making it robust to outliers.
- **Formula**:  
  $$[
  X_{\text{scaled}} = \frac{X - \text{Median}(X)}{\text{IQR}(X)}
  ]$$
- **Use When**:
  - Data contains outliers.
  - You want to minimize the influence of extreme values.

---

### ✅ Recommended Scaler for Your Case

Given that you've already performed outlier removal and are working with algorithms sensitive to feature scales (e.g., Logistic Regression, SVM), **StandardScaler** is recommended. It standardizes features by removing the mean and scaling to unit variance, which is suitable for algorithms that assume normally distributed data.

---

### 🛠️ Implementation Example

```python

# Fit on training data and transform

```

#### Encoding using **Label encoder**

**Label Encoding** is a method for converting categorical variables into numeric format by assigning each unique category a **distinct integer value**. For example, a column with values `["Red", "Green", "Blue"]` would be transformed into `[0, 1, 2]`. This technique is especially useful when the categorical variable has **ordinal (ordered)** relationships.
---

### **Importance of Label Encoding**

We use Label Encoding because:

1. **Efficiency and Simplicity**: Label Encoding is a memory-efficient way to convert categories to numbers, especially when dealing with features with many unique values (e.g., country names). Unlike One Hot Encoding, it doesn't increase dimensionality.

2. **Supports Ordinal Relationships**: For features where categories have a **natural order** (e.g., "Low", "Medium", "High"), Label Encoding preserves that order, making it more meaningful to some algorithms.

3. **Compatibility with Machine Learning Algorithms**: Many models (especially **tree-based algorithms** like Decision Trees and Random Forests) can handle label-encoded features well and don't misinterpret the numeric labels as ordinal unless the algorithm requires it.

> ⚠️ However, be cautious: **Label Encoding is not suitable for nominal (unordered) categorical variables** when used with models that assume ordinality (e.g., linear regression), as it may introduce bias.

[More about it](https://www.geeksforgeeks.org/ml-label-encoding-of-datasets-in-python/)

### Handlign our data imbalance

#### 📊 What Is Data Sampling in Machine Learning?

In machine learning, **data sampling** refers to techniques used to adjust the class distribution within a dataset. This is particularly crucial when dealing with **imbalanced datasets**, where one class (e.g., non-diabetic cases) significantly outnumbers another (e.g., diabetic cases). Such imbalance can lead models to be biased toward the majority class, resulting in poor performance on the minority class.

---

## ⚖️ Why Is Sampling Important?

Imbalanced datasets can cause several issues:

- **Biased Predictions**: Models may favor the majority class, neglecting the minority class.
- **Misleading Accuracy**: High overall accuracy might mask poor performance on the minority class.
- **Poor Generalization**: The model may not learn the characteristics of the minority class effectively.

To address these issues, sampling techniques are used to balance the class distribution and help the model learn more fairly.

---

## 🔄 Common Sampling Techniques

1. **Oversampling**: Increases the number of minority class samples.  
   - *Example*: **SMOTE (Synthetic Minority Over-sampling Technique)** generates synthetic samples by interpolating between existing ones.

2. **Undersampling**: Reduces the number of majority class samples.  
   - *Example*: Randomly removes samples from the majority class to create balance.

3. **Combination of Both**: Uses both oversampling and undersampling for more robust balance.  
   - *Example*: **SMOTETomek**, which combines SMOTE with Tomek Links.

---

## ✅ Why Did We Use SMOTETomek?

**SMOTETomek** is a hybrid technique that offers the benefits of both oversampling and undersampling:

- **SMOTE** creates synthetic samples to boost the minority class.
- **Tomek Links** remove overlapping or ambiguous samples near the decision boundary.

This approach improves class balance while also cleaning up noisy data, leading to better model generalization and performance.

---

## 📌 Summary of Benefits

Using SMOTETomek helps:

- ✅ Improve model performance, especially recall and precision on the minority class.
- ✅ Clean noisy or overlapping examples near decision boundaries.
- ✅ Ensure the model learns equally from both classes, avoiding majority bias.

---

### 🔄 Why Do We Use `fit_transform()` on Training Data But Only `transform()` on Test Data?

In machine learning preprocessing, it's standard to use `fit_transform()` on the **training data** and only `transform()` on the **test data**.

---

### 🔍 What Does `fit_transform()` Do?

- `fit()` computes the necessary parameters (e.g., mean and standard deviation for scalers, label mappings for encoders) from the training data.
- `transform()` applies these parameters to modify the data accordingly.
- `fit_transform()` is simply a shortcut for calling `fit()` and then `transform()` on the same data.

---

### 🚫 Why Not Use `fit_transform()` on Test Data?

Using `fit_transform()` on test data **recalculates** parameters based on that test data. This leads to **data leakage**, where information from the test set influences the training process.

#### ❗ Problems with data leakage:
- Inflated model performance
- Poor generalization to real-world data
- Invalid evaluation metrics

---

### ✅ Correct Approach

| Dataset      | Method           | Purpose                                  |
|--------------|------------------|------------------------------------------|
| Training set | `fit_transform()` | Learn parameters + apply transformation |
| Test set     | `transform()`     | Apply learned parameters only           |

This ensures a clean separation between model training and evaluation.

---

### 🧠 Analogy

Think of `fit()` as **studying for an exam**, and `transform()` as **taking the exam**.  
If you study using the test questions themselves (i.e., `fit_transform()` on test data), your results won't reflect true performance — it's like cheating.

---

### 🔍 Overview of Classification Models Used

This project compares the performance of several machine learning classification models. Each model has its own strengths and assumptions, making them suitable for different types of data distributions and problem settings.

---

### 1. Logistic Regression

**Purpose**: A linear model used for binary classification. It predicts the probability that an instance belongs to a particular class.

**Equation**:

$$
P(y = 1 \mid x) = \sigma(w^T x + b) = \frac{1}{1 + e^{-(w^T x + b)}}
$$

Where:  
- $( \sigma )$ is the sigmoid function  
- $( w )$ is the weights vector  
- $( b )$ is the bias term

**Strengths**: Simple, interpretable, works well with linearly separable data.

---

### 2. Random Forest

**Purpose**: An ensemble of decision trees that combines multiple trees trained on random subsets of the data and features.

**How It Works**:
- Each tree votes for a class.
- The majority vote is taken as the final prediction.

**Strengths**: Handles high-dimensional data well, reduces overfitting through averaging.

---

### 3. XGBoost (Extreme Gradient Boosting)

**Purpose**: A powerful gradient boosting algorithm that builds trees sequentially to correct the errors of previous trees.

**Core Idea**:
- Uses gradient descent to minimize a loss function.
- Regularization is built in to prevent overfitting.

**Objective Function**:

$$
\mathcal{L} = \sum_{i=1}^{n} l(y_i, \hat{y}_i^{(t)}) + \sum_{k=1}^{t} \Omega(f_k)
$$

Where $( \Omega(f_k) )$ penalizes the complexity of each tree.

**Strengths**: Fast, efficient, excellent performance on structured/tabular data.

---

### 4. Decision Tree

**Purpose**: A tree-like structure where each internal node represents a decision rule, and each leaf node represents a class label.

**How It Works**:
- Splits data based on feature thresholds to maximize information gain or reduce Gini impurity.

**Gini Impurity Equation**:

$$
G = 1 - \sum_{i=1}^{C} p_i^2
$$

Where $( p_i )$ is the probability of class $( i )$ at a node.

**Strengths**: Easy to visualize, interpretable, handles both numerical and categorical data.

---

### 5. Naive Bayes (Gaussian)

**Purpose**: A probabilistic model based on Bayes' Theorem with the assumption of feature independence.

**Bayes' Theorem**:

$$
P(y \mid x) = \frac{P(x \mid y) P(y)}{P(x)}
$$

For Gaussian NB, it assumes:

$$
P(x_i \mid y) \sim \mathcal{N}(\mu_{y}, \sigma_{y}^2)
$$

**Strengths**: Fast, works well with text data and high-dimensional inputs.

---

### 6. K-Nearest Neighbors (KNN)

**Purpose**: A non-parametric algorithm that classifies a sample based on the majority label among its $( k )$ nearest neighbors.

**Distance Metric (Euclidean)**:

$$
d(x, x_i) = \sqrt{ \sum_{j=1}^{n} (x_j - x_{ij})^2 }
$$

**Strengths**: Simple, effective for low-dimensional and well-separated data.

---

These models provide a diverse set of approaches—linear, tree-based, probabilistic, and instance-based—allowing for a well-rounded evaluation on the classification task.

## Ensemble Learning Algorithm

### 📘 Introduction to Ensemble Learning

Ensemble learning is a powerful machine learning paradigm where multiple models, often referred to as `"weak learners,"` are trained to solve the same problem and combined to obtain better predictive performance. The fundamental idea is that by aggregating the predictions of several models, the ensemble can achieve higher accuracy and robustness compared to any individual model.

### 🔍 Voting Classifier

One of the simplest and most commonly used ensemble methods is the **Voting Classifier**. This approach involves training multiple different models (e.g., logistic regression, decision trees, support vector machines) on the same dataset and then combining their predictions.

There are two main types of voting strategies:

- **Hard Voting**: Each classifier makes its own prediction (votes for a class), and the final prediction is the class that receives the majority of votes.

- **Soft Voting**: Each classifier predicts the probability of each class, and the final prediction is made by averaging these probabilities and selecting the class with the highest average probability.

The Voting Classifier is particularly useful when combining models that are individually strong in different aspects, as it leverages the strengths of each to improve overall performance.

### 🛠️ Implementation in Scikit-Learn

In Python's scikit-learn library, the `VotingClassifier` class allows for easy implementation of this ensemble method. Here's a brief example:

```python
from sklearn.ensemble import VotingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC

# Instantiate the individual models
model1 = LogisticRegression()
model2 = DecisionTreeClassifier()
model3 = SVC(probability=True)

# Create the Voting Classifier
voting_clf = VotingClassifier(estimators=[
    ('lr', model1), ('dt', model2), ('svc', model3)],
    voting='soft')  # Use 'hard' for hard voting

# Fit the model
voting_clf.fit(X_train, y_train)

# Make predictions
predictions = voting_clf.predict(X_test)


#### Soft Voting `(Depends on the highest average possibility)`

````python
voting_model = VotingClassifier(
    estimators=[
        ('lr', LogisticRegression(max_iter=1000)),
        ('rf', RandomForestClassifier()),
        ('xgb', XGBClassifier(eval_metric='logloss'))
    ],
    voting='soft'
)

voting_model.fit(X_train_bal, y_train_bal)
y_pred_voting = voting_model.predict(X_test)

print("--- VotingClassifier ---")
print("Accuracy:", accuracy_score(y_test, y_pred_voting))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_voting))
print("Classification Report:\n", classification_report(y_test, y_pred_voting))
````

#### Hard Voting `(Depends on the majority of votes)`

````python
voting_model = VotingClassifier(
    estimators=[
        ('lr', LogisticRegression(max_iter=1000)),
        ('rf', RandomForestClassifier()),
        ('xgb', XGBClassifier(eval_metric='logloss'))
    ],
    voting='hard'
)

voting_model.fit(X_train_bal, y_train_bal)
y_pred_voting = voting_model.predict(X_test)

# Calculate metrics
acc = accuracy_score(y_test, y_pred_voting)
report = classification_report(y_test, y_pred_voting, output_dict=True)
cm = confusion_matrix(y_test, y_pred_voting)

# Add to model_scores dictionary
model_scores["Voting Classifier"] = {
    "model": voting_model,
    "y_pred": y_pred_voting,
    "accuracy": acc,
    "report": report,
    "confusion_matrix": cm
}

print("--- VotingClassifier ---")
print("Accuracy:", accuracy_score(y_test, y_pred_voting))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_voting))
print("Classification Report:\n", classification_report(y_test, y_pred_voting))
````

# Visualizing Confusion Matrices for All Trained Models


## 📊 What is a Confusion Matrix?

A **confusion matrix** is a performance evaluation table used for classification problems.  
It compares the predicted labels from your model to the actual true labels.

---

### 📦 Layout of the Confusion Matrix:

$$
\begin{array}{c|cc}
\text{Predicted} \backslash \text{Actual} & \text{0} & \text{1} \\
\hline
\text{0} & \text{TN} & \text{FN} \\
\text{1} & \text{FP} & \text{TP} \\
\end{array}
$$

- **TN (True Negative):** Model predicted 0, and the actual was 0.
- **TP (True Positive):** Model predicted 1, and the actual was 1.
- **FP (False Positive):** Model predicted 1, but the actual was 0 (false alarm).
- **FN (False Negative):** Model predicted 0, but the actual was 1 (missed it).

---

### ✅ Why It Matters:

The confusion matrix helps you calculate key metrics like:
- **Accuracy:** Overall correctness → `(TP + TN) / Total`
- **Precision:** Correctness of positive predictions → `TP / (TP + FP)`
- **Recall (Sensitivity):** Ability to find all actual positives → `TP / (TP + FN)`
- **F1-Score:** Balance between precision and recall → `2 * (Precision * Recall) / (Precision + Recall)`

Use it to understand not just how often your model is right — but **how** it is right or wrong.


````python
num_models = len(model_scores)
cols = 4
rows = -(-num_models // cols)  # ceiling division without importing math module

fig, axes = plt.subplots(rows, cols, figsize=(15, 5 * rows))
axes = axes.flatten()

for idx, (name, info) in enumerate(model_scores.items()):
    sns.heatmap(info["confusion_matrix"], annot=True, fmt='d', cmap='Blues', cbar=False, ax=axes[idx])
    axes[idx].set_title(f"Confusion Matrix - {name}")
    axes[idx].set_xlabel("Predicted")
    axes[idx].set_ylabel("Actual")

# Turn off any unused subplots
for i in range(idx + 1, len(axes)):
    axes[i].axis('off')

plt.tight_layout()
plt.show()
````
